```python
import matplotlib.pyplot as plt

# Movie rating categories and counts
rating_categories = ['Poor (0-2)', 'Average (2-3)', 'Good (3-4)', 'Excellent (4-5)']
movie_counts = [450, 3200, 18500, 12700]

# Custom colors
colors = ['#ff6b6b', '#feca57', '#48dbfb', '#1dd1a1']

plt.figure(figsize=(8,5))
bars = plt.bar(rating_categories, movie_counts, color=colors)

plt.xlabel("Rating Category")
plt.ylabel("Number of Movies")
plt.title("Distribution of Movie Ratings")

# Display values on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2,
             height,
             f'{int(height)}',
             ha='center',
             va='bottom')

plt.tight_layout()
plt.show()
```
