```python
import matplotlib.pyplot as plt

rating_categories = ['Poor (0-2)', 'Average (2-3)', 'Good (3-4)', 'Excellent (4-5)']
movie_counts = [450, 3200, 18500, 12700]

# New color palette
colors = ['#6c5ce7', '#74b9ff', '#0984e3', '#00cec9']

plt.figure(figsize=(8,5))
bars = plt.bar(rating_categories, movie_counts, color=colors)

plt.xlabel("Rating Category")
plt.ylabel("Number of Movies")
plt.title("Distribution of Movie Ratings")

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2,
             height,
             f'{int(height)}',
             ha='center',
             va='bottom')

plt.tight_layout()
plt.show()
```
